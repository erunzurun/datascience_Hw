```python
test_1 = np.array([92,94,88,91,87])
test_2 = np.array([79,100,86,93,91])
test_3 = np.array([87,85,72,90,92])

test_3_fix = test_3 + 2
print(test_3_fix)

total_grade = test_1 + test_2 + test_3_fix
print(total_grade)

final_grade = total_grade / 3
print(final_grade)

student_scores = np.array([test_1, test_2, test_3_fix])
print(student_scores.shape)
```

```python
jeremy_test_2 = student_scores[1,3]
print(jeremy_test_2)

m_a_test_1 = np.array([student_scores[0,1], student_scores[0, 2]])
print(m_a_test_1)

print(student_scores[:,4])
print(student_scores[:,-1])
```

```python
a = np.array([10,2,2,4,5,3,9,8,9,7])
mask = a>5
print(mask)

selected = a[a>5]
print(selected)

selected2 = a[(a>5) | (a<4)]
print(selected2)
```

```python
porridge = np.array([79, 65, 50, 63,56,90,85,98,79,51])

cold = np.array(porridge[(porridge<60)])

hot = np.array(porridge[(porridge>80)])

just_right = np.array(porridge[(porridge>60) & (porridge<80)])

print(cold)
print(hot)
print(just_right)
```
