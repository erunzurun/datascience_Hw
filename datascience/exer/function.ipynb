```python
def check_leap_year(year):
    if year % 4 == 0:
        return "%s is a leap year"%(year)
    else:
        return "%s is not a leap year"%(year)
        
check_leap_year(2021)       
```

```python
def square_point(x, y, z):
    x_squared = x**2
    y_squared = y**2
    z_squared = z**2
    return x_squared, y_squared ,z_squared

three, four, five = square_point(3,4,5)
print(three, four, five)
```

```python
def findvolume(h, w, d=3):
    print("Height = %s" % (h))
    print("Width = %s" % (w))
    print("Depth = %s" % d)
    return h*w*d
    
print(findvolume(w = 2,h = 1,d = 4))
```
