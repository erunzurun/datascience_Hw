```python
np.random.seed(0)

a = np.random.randint(10, size = (3,4))
print(a)

print(a < 6)
```

```python
print(a)
print(np.count_nonzero(a<6)) # false : 0
# print(np.sum)

print(np.any(a>8))
print(np.all(a>1))
```

```python
np.random.seed(1)
x = np.random.randint(100, size = 10)
print(x)

a = x[2]; b = x[4]; c = x[6];

print(a, b, c)

ind = [2,4,6]
print(x[ind])
```

```python
np.random.seed(2)
x = np.random.randint(10, size = (3,5))
print(x)

row = [0, 1, 2]
col = [2, 1, 4]

print(x[row, col])
x[row, col]
```

```python
print(x)

row = np.array([0,1,2])

col = np.array([2,1,4])

x[row[:, np.newaxis], col]
```
